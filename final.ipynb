```python
import cv2 
  
```

```python
def FrameCapture(path): 
      
    # Path to video file 
    vidObj = cv2.VideoCapture(path) 
  
    # Used as counter variable 
    count = 0
  
    # checks whether frames were extracted 
    success = 1
    flg=0
    while success:  
  
        # vidObj object calls read 
        # function extract frames 
        success, image = vidObj.read() 
        if count % 7 ==0:
            # Saves the frames with frame-count 
            #cv2.imwrite("./apu/frame%d.jpg" % count, image) 
            face_cascade = cv2.CascadeClassifier('/root/anaconda3/share/OpenCV/haarcascades/haarcascade_frontalface_default.xml') 

            # https://github.com/Itseez/opencv/blob/master 
            # /data/haarcascades/haarcascade_eye.xml 
            # Trained XML file for detecting eyes 
            eye_cascade = cv2.CascadeClassifier('/root/anaconda3/share/OpenCV/haarcascades/haarcascade_eye.xml')  

            img = image

            # convert to gray scale of each frames 
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

            # Detects faces of different sizes in the input image 
            faces = face_cascade.detectMultiScale(gray, 1.3, 5) 
            
            for (x,y,w,h) in faces:
                flg=flg+1
                if(len(faces)>2):
                    break
                height = img.shape[0]
                width = img.shape[1] 
                x=max(0,x-int(w/8))
                w=3*int(w/2.5)
                y=max(0,y-int(h/3))
                h=int(h*1.5)
                
               # print(x,y,w,h)
                crop_img = img[y:min(height,y+h), x:min(width,x+w)]
                tergetwidth = 300
                tergetheight = 300
                dim = (tergetwidth, tergetheight)

                # resize image
                resized_final_sample_img = cv2.resize(crop_img, dim, interpolation = cv2.INTER_AREA) 
                cv2.imwrite("./datasetss/person4/"+ str(flg)+"frame%d .jpg" % count, resized_final_sample_img) 
                #cv2.imwrite("./testfold1/frame%dfcs.jpg" % count, image) 
                print(count)
            
  
        
  
        count += 1
    
  
# Driver Code 
if __name__ == '__main__': 
  
    # Calling the function 
    print("apu")
    FrameCapture("./data set/person4.mp4") 
```
